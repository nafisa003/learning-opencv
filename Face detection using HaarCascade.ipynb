```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv.imread('test/lilly.webp')
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray,cmap="gray")
```

```python
haar_cascade=cv.CascadeClassifier('haar_face.xml')
```

```python
faces_rect=haar_cascade.detectMultiScale(gray,scaleFactor=1.4,minNeighbors=1)
#detects face and returns the rectangular coordinates as a list

print(len(faces_rect)) #indicates no of faces found
```

```python
for (x,y,w,h) in faces_rect:
    cv.rectangle(img,(x,y),(x+w,y+h),(0,0,255),3)
    
# img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
plt.imshow(img)
```
