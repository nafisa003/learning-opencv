```python
blank=np.zeros((500,500,3),dtype='uint8')
```

```python
# cv.imshow('blank',blank)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank[200:300,100:200]=255,0,0 ##RGB
plt.imshow(blank)
plt.show()
```

```python
blank=cv.rectangle(blank,(0,0),(250,250),(0,255,0),thickness=5)
plt.imshow(blank)
```

```python
blank=cv.rectangle(blank,(0,0),(blank.shape[1]//2,blank.shape[0]//2),(0,255,0),thickness=-1)
plt.imshow(blank)
```

```python
##draw a circle
circle=cv.circle(blank,(blank.shape[1]//2,blank.shape[0]//2), 100,(0,0,255),thickness=3)
plt.imshow(circle)
```

```python
new_circle=cv.circle(blank,(blank.shape[1]//2,blank.shape[0]//2), 100,(0,0,255),thickness=-1)
plt.imshow(new_circle)
```

```python
##draw a line
cv.line(blank,(0,0),(blank.shape[1]//2,blank.shape[0]//2), (0,0,0),thickness=10)
plt.imshow(blank)
cv.line(blank,(100,400),(500,800), (0,0,0),thickness=10)
plt.imshow(blank)
```

# Basic functions in opencv

```python
##converting to grayscale to observe intensity distriution of pixels
```

```python
img=cv.imread('test/maria.jpeg')
```

```python
plt.imshow(img)
```

```python
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray",vmin=0,vmax=255)
```

```python
##blurring an image to reduce noise due to issues with camera,sensor or so on using gaussian blur
```

```python
blur=cv.GaussianBlur(img,(5,5),cv.BORDER_DEFAULT)##incresig kernel size will increase blur (5,5)->(7,7)
plt.imshow(blur)
```

```python
##edge cascade
```

```python
edges=cv.Canny(img,120,175)
plt.imshow(edges,cmap="gray")
```

```python
edges=cv.Canny(blur,120,175) ##we can reduce edges by applying blur
plt.imshow(edges,cmap="gray")
```

```python
##dilating an image using structuring element (here canny edges)
```

```python
dilated=cv.dilate(edges,(3,3),iterations=1)
plt.imshow(dilated,cmap="gray")##we get dilated edges
```

```python
###eroding the image to get back structure element
```

```python
eroded=cv.erode(dilated,(3,3),iterations=1)
plt.imshow(eroded,cmap="gray")
```

```python
##resizing an image
```

```python
resized=cv.resize(img,(500,500)) ##ignores the aspect ratio
plt.imshow(resized)
plt.show()
```

```python
##interpolation (INTER_AREA) are useful if we are shrinking the image from its original dimensions
##to enlarge dimensions INTER_LINEAR, INTER_CUBIC (slower) useful
resized2=cv.resize(img,(5000,3000),interpolation=cv.INTER_LINEAR) ##ignores the aspect ratio
plt.imshow(resized2)
```
