```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
haar_cascade=cv.CascadeClassifier('haar_face.xml')
```

```python
features=np.load('features.npy',allow_pickle=True)
labels=np.load('labels.npy')
```

```python
face_recognizer=cv.face.LBPHFaceRecognizer_create()
face_recognizer.read('face_recognizer_trained.yml')
```

```python
people=[]
DIR=r'C:\Users\USER\Documents\assignments\dataset'
for person in os.listdir(DIR):
    people.append(person)
people
    
```

```python
#test image
img=cv.imread('test/fed.jpg')
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
```

```python
faces_rect=haar_cascade.detectMultiScale(gray,1.1,4)
```

```python
for (x,y,w,h) in faces_rect:
    faces_roi=gray[y:y+h,x:x+h]
    label,confidence=face_recognizer.predict(faces_roi)
    print(f'Label={people[label]} with a confidence of {confidence}')
    
    cv.putText(img,str(people[label]),(20,20), cv.FONT_HERSHEY_COMPLEX, 1.0, (0,255,0),2)
    cv.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
plt.imshow(img)
    
```
