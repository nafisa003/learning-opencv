```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv.imread('test/maria.jpeg')
```

```python
plt.imshow(img)
```

```python
##translating an image
```

```python
def translate(img,x,y): #x and y are number of pixels to be shifted along x and y axis
    transMat=np.float32([[1,0,x],[0,1,y]]) ##creating transaltion matrix
    dimensions=(img.shape[1],img.shape[0])
    return cv.warpAffine(img,transMat,dimensions)
# -x-->left
# x-->right
# -y-->up
# y-->down

# What is an Affine Transformation?
# A transformation that can be expressed in the form of a matrix multiplication (linear transformation) followed by a vector addition (translation).
# From the above, we can use an Affine Transformation to express:

# Rotations (linear transformation)
# Translations (vector addition)
# Scale operations (linear transformation)
# you can see that, in essence, an Affine Transformation represents a relation between two images.


# We use the function cv::warpAffine for that purpose.
    
```

```python
translated=translate(img,100,100)
plt.imshow(translated)
```

```python
##rotation
```

```python
def rotate(img,angle,rotPoint=None):
    (height,width)=img.shape[:2]
    
    if rotPoint==None:
        rotPoint=(width//2,height//2)#center
    
        
    rotMat=cv.getRotationMatrix2D(rotPoint,angle,scale=1.0)
    dimensions=(width,height)
    return cv.warpAffine(img,rotMat,dimensions)
```

```python
rotated=rotate(img,45)
plt.imshow(rotated)
```

```python
rotated2=rotate(img,-45) #rotation clockwise
plt.imshow(rotated2)
```

```python
rotated3=rotate(rotated2,-45) #rotation clockwise
plt.imshow(rotated3) 
```

##we get a skewed output due to rotation of the black triangles along with the rotated image so we can just add up the rotation angles to avoid this issue

```python
rotated4=rotate(img,-90) #rotation clockwise
plt.imshow(rotated4)
```
