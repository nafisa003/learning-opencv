```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv.imread('test/maria.jpeg')
plt.imshow(img)
```

```python
##BGR TO GRAY
##grayscale images show the pixel intensities at particular locations of the image

gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
```

```python
##BGR TO HSV(huge saturation value)
hsv=cv.cvtColor(img,cv.COLOR_BGR2HSV)
plt.imshow(hsv,cmap="hsv")
```

```python
##BGR TO Lab
lab=cv.cvtColor(img,cv.COLOR_BGR2LAB)
plt.imshow(lab)
```

```python
rgb=cv.cvtColor(img,cv.COLOR_BGR2RGB) ##opencv reads in BGR and matplotlib in RGB format
plt.imshow(rgb)
```

```python
##grayscale cannot be directly converted to hsv first need to convert to bgr
##same goes for lab
hsv_bgr=cv.cvtColor(hsv,cv.COLOR_HSV2BGR)
gray_from_bgr=cv.cvtColor(hsv_bgr,cv.COLOR_BGR2GRAY)
plt.imshow(gray_from_bgr,cmap="gray")
```

# Split and merge color channels

```python
b,g,r=cv.split(img)
print(r.shape)
plt.imshow(r,cmap="gray")
```

```python
print(b.shape)
plt.imshow(b,cmap="gray")
```

```python
print(g.shape)
plt.imshow(g,cmap="gray")
```

darker region represents there are very less pixels of that color in that region and lighter region indicates more pixel values of that color

```python
merged=cv.merge([r,g,b])
plt.imshow(merged)
```

```python
print(g.shape)
plt.imshow(g,cmap="Greens")
```

```python
b,g,r=cv.split(img)
blank=np.zeros(img.shape[:2],dtype="uint8")
red=cv.merge([r,blank,blank])
green=cv.merge([blank,g,blank])
blue=cv.merge([blank,blank,b])
```

```python



fig = plt.figure(figsize=(20, 10))
rows = 1
columns = 3

fig.add_subplot(rows, columns, 1)
  
# showing image
plt.imshow(blue)
plt.axis('off')
plt.title("Blue")
  
# Adds a subplot at the 2nd position
fig.add_subplot(rows, columns, 2)
  
# showing image
plt.imshow(green)
plt.axis('off')
plt.title("Green")
  
# Adds a subplot at the 3rd position
fig.add_subplot(rows, columns, 3)
  
# showing image
plt.imshow(red)
plt.axis('off')
plt.title("Red")

plt.show()
```

# Blurring

```python
img=cv.imread('test/maria.jpeg')
plt.imshow(img)
```

```python
##smoothing and blurring is needed to reduce the noise in images
```

```python
#averaging
avg=cv.blur(img,(3,3))##increasing kernel size increases blur
plt.imshow(avg)
```

```python
#Gaussian
gau=cv.GaussianBlur(img,(3,3),0)
plt.imshow(gau)#less blur than averaging
```

```python
#median
##removes substantial noise more than avg and gaussian
##used in advanced computer vision projects

median=cv.medianBlur(img,3)
plt.imshow(median)
```

```python
#bilateral
#most effective and used in lot of projects in CV
#retains edges of image

bilateral=cv.bilateralFilter(img,d=10, sigmaColor=50, sigmaSpace=25)
# higher sigma color means more colors in neighborhood are cosidered when blur is computed
# higher sigma space means that pixels further out from central pixel will influence blur calculation
plt.imshow(bilateral)
```
