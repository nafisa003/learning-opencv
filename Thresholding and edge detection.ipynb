```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv.imread('test/maria.jpeg')
gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
plt.imshow(gray,cmap="gray")
```

```python
#simple thresolding
```

```python
threshold,thresh=cv.threshold(gray,100,255,cv.THRESH_BINARY) 

#compares each pixel value to the threshold 
#pixels less than 150 is set to black
#pixels greater than 150 to 255 or white
#return thresh (thresholded image) and threshold which is 150 that is passed

plt.imshow(thresh,cmap="gray")
```

```python
threshold,thresh=cv.threshold(gray,100,255,cv.THRESH_BINARY_INV)
plt.imshow(thresh,cmap="gray")

#change less than 150 to white and vice versa
```

```python
#adaptive thresholding
```

```python
adaptive_thresh=cv.adaptiveThreshold(gray,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY, 13, 3)

#adaptive threshold mean is the strategy or method that takes mean of neighboring pixels
# 10 is the number of blocksize or kernel size to calculate mean and has to be odd
# C=3 is an integer subtracted from the mean aloowing to fine tune the threshold
```

```python
plt.imshow(adaptive_thresh,cmap="gray")
```

```python
adaptive_thresh=cv.adaptiveThreshold(gray,255,cv.ADAPTIVE_THRESH_MEAN_C,cv.THRESH_BINARY, 23, 5)
plt.imshow(adaptive_thresh,cmap="gray")
```

```python
adaptive_thresh=cv.adaptiveThreshold(gray,255,cv.ADAPTIVE_THRESH_GAUSSIAN_C,cv.THRESH_BINARY, 23, 5)
plt.imshow(adaptive_thresh,cmap="gray")
```

```python
#better image than mean
```

```python
# Gradients and Edge detection
```

```python
#Laplacian
```

```python
lap=cv.Laplacian(gray,cv.CV_64F) #calculates gradients having + and - slopes
lap=np.uint8(np.absolute(lap)) #pixels can't be negative so taking absolute
plt.imshow(lap,cmap="gray")

#detects laplacian edges
```

```python
#Sobel

#computes gradients along x and y axis 
```

```python
sobelx=cv.Sobel(gray,cv.CV_64F, 1 ,0)
sobely=cv.Sobel(gray,cv.CV_64F, 0 ,1)

fig=plt.figure(figsize=(20,10))

rows=1
cols=2

fig.add_subplot(rows,cols,1)
plt.imshow(sobelx,cmap='gray')
plt.title("SobelX")

fig.add_subplot(rows,cols,2)
plt.imshow(sobely,cmap='gray')
plt.title("SobelY")

plt.show()

```

```python
combined_sobel=cv.bitwise_or(sobely,sobelx)
plt.imshow(combined_sobel,cmap="gray")
```

```python
canny=cv.Canny(gray,150,175)
plt.imshow(canny,cmap="gray")
```

```python
#canny is the mostly used and advanced method, it uses sobel as one of its steps to calculate gradients

#many advanced cases use sobel
```
