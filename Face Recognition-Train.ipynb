```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
people=[]
for i in os.listdir(r'C:\Users\USER\Documents\assignments\dataset'):
    people.append(i)
people
```

```python
DIR=r'C:\Users\USER\Documents\assignments\dataset'
```

```python
features=[]
labels=[]
haar_cascade=cv.CascadeClassifier('haar_face.xml')
```

```python
#create training set
def create_train():
    for person in people:
        path=os.path.join(DIR,person)
        label=people.index(person)
        
        for img in os.listdir(path):
            img_path=os.path.join(path,img)
            
            img_array=cv.imread(img_path)
            
            print(img_path)
            gray=cv.cvtColor(img_array,cv.COLOR_BGR2GRAY)
            
            faces_rect=haar_cascade.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=4)
            
            for (x,y,w,h) in faces_rect:
                faces_roi=gray[y:y+h,x:x+w] #finding region of interest for faces in image
                features.append(faces_roi)
                labels.append(label)
create_train()
print(len(features))
print(len(labels))
```

```python
print(len(features))
print(len(labels))
face_recognizer=cv.face.LBPHFaceRecognizer_create()
#creating an instance of face recognizer
```

```python
#train thr recognizer on the features and labels list
features=np.array(features,dtype='object')
labels=np.array(labels)
face_recognizer.train(features,labels)
```

```python
face_recognizer.save('face_recognizer_trained.yml') #saving the trained model
```

```python
np.save('features.npy',features)
np.save('labels.npy',labels)
```
