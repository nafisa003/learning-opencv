```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

# Bitwise operations

```python
blank=np.zeros((400,400),dtype="uint8")
rect=cv.rectangle(blank.copy(),(30,30),(370,370),1,-1)
circle=cv.circle(blank.copy(),(200,200),200,1,-1)

fig=plt.figure(figsize=(10,6))
rows=1
cols=2

fig.add_subplot(1,2,1)
plt.imshow(rect,cmap="gray")
fig.add_subplot(1,2,2)
plt.imshow(circle,cmap="gray")

plt.show()
```

```python
bitwise_and=cv.bitwise_and(rect,circle)
plt.imshow(bitwise_and,cmap='gray') ##intersection of set
```

```python
bitwise_or=cv.bitwise_or(rect,circle)
plt.imshow(bitwise_or,cmap='gray') ##union of set
```

```python
bitwise_xor=cv.bitwise_xor(rect,circle)
plt.imshow(bitwise_xor,cmap='gray') ##non-intersecting regions
```

```python
bitwise_not=cv.bitwise_not(rect)
plt.imshow(bitwise_not,cmap='gray') ##inverts the color
```

# Masking

```python
##masking allows us to focus on certain parts of image that we are interested in
```

```python
img=cv.imread('test/maria.jpeg')
```

```python
blank=np.zeros(img.shape[:2],dtype="uint8")
```

```python
mask=cv.circle(blank.copy(),(img.shape[1]//2,img.shape[1]//2),30,255,-1)
```

```python
plt.imshow(mask,cmap="gray")
```

```python
masked_image=cv.bitwise_and(img,img,mask=mask)
plt.imshow(masked_image)
```

```python
mask2=cv.rectangle(blank.copy(),(img.shape[1]//2,img.shape[1]//2),(img.shape[1]//2+100,img.shape[1]//2+100),255,-1)
masked_image2=cv.bitwise_and(img,img,mask=mask2)
plt.imshow(masked_image2)
```

```python
blank=np.zeros(img.shape[:2],dtype="uint8")
rect=cv.rectangle(blank.copy(),(30,30),(370,370),1,-1)
circle=cv.circle(blank.copy(),(200,200),200,1,-1)

weird_shape=cv.bitwise_and(rect,circle)
masked_image3=cv.bitwise_and(img,img,mask=weird_shape)
plt.imshow(masked_image3)
```
